### 7장 퍼셉트론

#### (1) AND Gate

```python
## AND 퍼셉트론
import numpy as np
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y_AND = np.array([[0], [0], [0], [1]])

def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))

W = [[1], [1]]  # 임의의 가중치 (초기화)
b = -1.5
Z = np.matmul(X, W) + b
Y_hat = sigmoid(Z)
Y_hat
```

```python
(Y_hat > 0.5).astype('i')
```

```python
## AND Gate :  X1, X2 평면으로 시각화

import matplotlib.pyplot as plt
import numpy as np

## 모든 좌표에서 색상 지정 Scatter Plot
for i in range(len(X)):
    if Y_AND[i] == 1:
        c="orange"      # True
        label = "True"
    else:
        c = 'blue'      # False
        label = "False"
    plt.scatter(X[i, 0], X[i, 1], s=50, c=c, label=label)
    
## X1, X2 평면에 직선 그리기
W1, W2, b = 1, 1, -1.5
X1 = np.arange(-1, 2, 0.1)
X2 = -(W1 / W2)*X1 - (b / W2)

plt.plot(X1, X2, "r")
plt.xlim([-0.5, 1.5])
plt.ylim([-0.5, 1.5])
plt.legend()
plt.grid()
```

```python
## Perceptron 시각화 함수 정의

import matplotlib.pyplot as plt
import numpy as np

def perceptron_plot(X, Y, W1, W2, b):
    for i in range(len(X)):
        if Y[i] == 1:
            c="orange"  # True
            label = "True"
        else:
            c = 'blue'  # False
            label = "False"
        plt.scatter(X[i, 0], X[i, 1], s=50, c=c, label=label)
    
    ## X1, X2 평면에 직선 그리기
    X1 = np.arange(-1, 2, 0.1)
    X2 = -(W1 / W2)*X1 - (b / W2)

    plt.plot(X1, X2, "r")
    plt.xlim([-0.5, 1.5])
    plt.ylim([-0.5, 1.5])
    plt.legend()
    plt.grid()
    plt.show()
```

```python
## AND Gate
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y_AND = np.array([[0], [0], [0], [1]])
W1, W2, b = 1, 1, -1.5
perceptron_plot(X, Y_AND, W1, W2, b)
```

### (2) OR Gate

```python
## OR Gate
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y_OR = np.array([[0], [1], [1], [1]])
W1, W2, b = 1, 1, -0.5
perceptron_plot(X, Y_OR, W1, W2, b)
```

#### (3) NAND Gate

```python
## NAND Gate
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y_NAND = np.array([[1], [1], [1], [0]])
W1, W2, b = -1, -1, 1.5
perceptron_plot(X, Y_NAND, W1, W2, b)
```

```python
# AND GATE

def AND(x1, x2):
    w1, w2, b = 0.5, 0.5, -0.7
    y = x1 * w1 + x2 * w2 + b
    if y <= 0:  # 임계치
        return 0
    else:
        return 1
    
print(AND(0, 0))
print(AND(1, 0))
print(AND(0, 1))
print(AND(1, 1))
```

```python
# NAND GATE

def NAND(x1, x2):
    w1, w2, b = -0.5, -0.5, 0.7
    y = x1 * w1 + x2 * w2 + b
    if y <= 0:  # 임계치
        return 0
    else:
        return 1
    
print(NAND(0, 0))
print(NAND(1, 0))
print(NAND(0, 1))
print(NAND(1, 1))
```

```python
# OR Gate

def OR(x1, x2):
    w1, w2, b = 0.5, 0.5, -0.25
    y = x1 * w1 + x2 * w2 + b
    if y <= 0:  # 임계치
        return 0
    else:
        return 1
    
print(OR(0, 0))
print(OR(1, 0))
print(OR(0, 1))
print(OR(1, 1))
```

```python
### XOR Gate!!!

def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y

print(XOR(0, 0))
print(XOR(1, 0))
print(XOR(0, 1))
print(XOR(1, 1))
```

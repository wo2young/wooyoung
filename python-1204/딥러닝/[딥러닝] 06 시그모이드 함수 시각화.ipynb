### 시그모이드 함수 시각화  
 
시그모이드 함수 :  $$y = \frac{1}{1 + e^{-(ax + b)}}$$


```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 불러오기

X = np.arange(-10, 10, 0.3)
Y = 1/(1 + np.exp(-X))

plt.yticks([0,0.5,1])
plt.grid()
plt.plot(X,Y, color='red')
plt.scatter(X,Y)
plt.show()
```

```python
# 데이터 불러오기
a,b = 2, -5
X = np.arange(-10, 10, 0.3)
Y = 1/(1 + np.exp(-a * X + b))

plt.yticks([0,0.5,1])
plt.grid()
plt.plot(X,Y, color='red')
plt.show()
```

```python
a, b = -1, 0
X = np.arange(-10, 10, 0.1)
Y = 1 / (1 + np.exp(-X))   # 기본 시그모이드
Y1 = 1 / (1 + np.exp(-(a* X + b)))
plt.xticks([-10, -8, -6, -4, -2, 0, 2, 4, 6, 8, 10])
plt.yticks([0, 0.25, 0.5, 0.75, 1])
plt.plot(X, Y, c="g")
plt.plot(X, Y1, c="r")
plt.grid()
plt.show()
```

```python
def sigmoid(T):
    return 1 / (1 + np.exp(-T))

def plot_sigmoid(X, Y, color):
    plt.plot(X, Y, c=color)
    plt.grid()
    plt.xticks([-10, -8, -6, -4, -2, 0, 2, 4, 6, 8, 10])
    plt.yticks([0, 0.25, 0.5, 0.75, 1])
    
    
a, b = 1, 0
plot_sigmoid(X, sigmoid(a * X + b), "k")  # black

# a, b = 1,  3
# plot_sigmoid(X, sigmoid(a * X + b), "g") # green

# a, b = 1,  -3
# plot_sigmoid(X, sigmoid(a * X + b), "r") # red

a, b = 2,  0
plot_sigmoid(X, sigmoid(a * X + b), "orange") # orange

# a, b = 0.5,  0
# plot_sigmoid(X, sigmoid(a * X + b), "y") # yellow

a, b = -1,  0
plot_sigmoid(X, sigmoid(a * X + b), "b") # blue

a, b = -1,  2
plot_sigmoid(X, sigmoid(a * X + b), "r") # red
```

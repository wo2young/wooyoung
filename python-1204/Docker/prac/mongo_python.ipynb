```python
import pymongo

# ==========================================
# [설정] 접속 정보
# ==========================================
# 1. 윈도우(Local) 실행 시: (기본값)
DB_HOST = "127.0.0.1"

# 2. Docker(Ubuntu) 컨테이너 내부 실행 시: (주석 해제 후 사용)
# DB_HOST = "host.docker.internal"

# Docker 실행 시 설정한 포트 (-p 27018:27017)
DB_PORT = 27018 
DB_PASS = "pass"

# ------------------------------------------

# 1. 연결 URL 생성
mongo_url = f"mongodb://root:{DB_PASS}@{DB_HOST}:{DB_PORT}/"

# 2. 연결
client = pymongo.MongoClient(mongo_url)

# 연결 테스트
client.admin.command('ping')
print(f"✅ Connected to MongoDB at {DB_HOST}:{DB_PORT}")

# 3. DB 및 컬렉션 선택
db = client["lecture_db"]
col = db["logs"]

# 4. 데이터 삽입
log_data = {
    "level": "INFO", 
    "msg": "Server started", 
    "tags": ["aws", "linux"]
}
res = col.insert_one(log_data)
print(f"Saved ID: {res.inserted_id}")

# 5. 조회
print(list(col.find({}, {"_id": 0})))
```

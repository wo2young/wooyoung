```python
import pandas as pd
from sqlalchemy import create_engine

# ==========================================
# [설정] 접속 정보
# ==========================================
# 윈도우(Local) 실행 시 127.0.0.1 사용 권장
DB_HOST = "127.0.0.1"

# 2. Docker(Ubuntu) 컨테이너 내부 실행 시: (주석 해제 후 사용)
# DB_HOST = "host.docker.internal"

# MySQL 포트 (Docker 설정값)
DB_PORT = 3307 
DB_PASS = "pass"

# ------------------------------------------

# 1. SQLAlchemy 엔진 생성
db_url = f"mysql+pymysql://root:{DB_PASS}@{DB_HOST}:{DB_PORT}/lecture_db"
engine = create_engine(db_url)

# 연결 테스트
with engine.connect() as connection:
    print(f"✅ SQLAlchemy Engine Connected to {DB_HOST}:{DB_PORT}")

# 2. 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2024-01-01', '2024-01-02', '2024-01-03'],
    'sales': [1000, 1500, 2000],
    'platform': ['Mobile', 'PC', 'Mobile']
})

# 3. DB에 저장
df.to_sql(name='daily_sales', con=engine, if_exists='replace', index=False)
print("DataFrame saved to DB!")

# 4. SQL 결과 -> DataFrame 변환
result_df = pd.read_sql("SELECT * FROM daily_sales", con=engine)
print(result_df)
```
